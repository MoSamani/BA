```python
import ast
import tldextract

#In diser Funktion wird die Liste von URLs durchgesucht und alle Domäne aus der URLs extrahiert.
#Die Domäne werden mit einem Leerzeichen von einander getrennt, gespeichert und als string zurückgegeben.

def get_domain(list_urls):
    l = ast.literal_eval(list_urls)
    domain_txt =''
    for x in range(0, len(l)):
        ex = tldextract.extract(str(l[x]))
        if x != 0:
            domain_txt = domain_txt + ' ' + ex.registered_domain
        else:
            domain_txt = domain_txt + ex.registered_domain
    if len(domain_txt) == 0:
        return '0'
    else :
        return domain_txt
```

```python
#Für alle Tweets wird die Methode gerufen
df_tweetskb_urls['URLs'] = df_tweetskb_urls['URLs'].astype(str)
df_tweetskb_urls['URLs_dom'] = df_tweetskb_urls.apply(lambda row: get_domain(row['URLs']), axis=1)
print('URLs_domain!')

```

```python
#Die String-Daten werden durch die Funktion split() geteilt, zu einer Liste umgewandelt.
#Die Länge der Liste wird  zurückgegeben

def get_domain_num(domains):
    l = domains.split()
    return len(l)
    
df_tweetskb_urls['URLs_dom'] = df_tweetskb_urls['URLs_dom'].astype(str)
df_tweetskb_urls['Anzahl der URLs'] = df_tweetskb_urls.apply(lambda row: get_domain_num(row['URLs_dom']) if (row['URLs_dom'] != '0') else 0, axis=1)
print('Anzahl der URLs wurde gerechnet!')
```

```python
#Labels und Pfad für den TweetsKB

TweetsKB_labels    = ['Tweet_Id', 'Benutzername', 'Zeitstempel', 'Abonnenten', 'Abonnierten', 'Retweets', 'Likes', 'Text', 'URLs', 'Ist_gelöscht', 'Hashtags', 'Erwähnungen', 'Anzahl der Hashtags', 'Anzahl der Mentions', 'Textlänge', 'Lesbarkeit' ]
TweetsKB_Pfad =  'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/TweetsKB_stichprobe_lesbarkeit.tsv'
```

```python
#TweetsKB-Stichprobe wird gelesen

df_tweetskb = pd.read_csv(TweetsKB_Pfad, sep='\t', error_bad_lines=False, names=TweetsKB_labels)
print('Form :', df_tweetskb.shape)
df_tweetskb.head()
```

```python
#Nur die TweetsIDs, Domäne und Anzahl der Domäne(URLs) werden genommen
df_tweetskb_urls = df_tweetskb_urls[['Tweet_Id', 'URLs_dom', 'Anzahl der URLs']]

#Hier werden die Domäne und Anzahl der URLs zu dem TweetsKB hinzugefügt. 
df_tweetskb = pd.merge(df_tweetskb, df_tweetskb_urls, on='Tweet_Id', how='left')
print('Form :', df_tweetskb.shape)
df_tweetskb.head()
```

```python
#Hier wird geprüft ob es Null-Daten in Dataframe gibt
df_tweetskb.isna().sum()
```

```python
#Dataframe wird gespeichert.

TweetsKB_Pfad =  'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/TweetsKB_stichprobe_domain_zahl.tsv'

with open(TweetsKB_Pfad, 'w', newline='', encoding="utf-8") as out_file:
    tsv_writer = csv.writer(out_file, delimiter="\t")
    for idx, row in df_tweetskb.iterrows():
        tsv_writer.writerow(row.values.tolist())
```

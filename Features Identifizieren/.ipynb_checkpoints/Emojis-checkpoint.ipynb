```python
#Hier list man extrahieren Tweets

pfad_tsv1 = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/random_sample/emoji/tweetskb_emoji.tsv'

df_tweetskb_emojies = pd.read_csv(pfad_tsv1, sep='\t', error_bad_lines=False, names=['Tweet_Id', "Emojie"])
print('Form :', df_tweetskb_emojies.shape)
df_tweetskb_emojies.head()
```

```python
#Hier wird die Anzahl der Emojis berechnet.
#Zuerst wird String-Daten zu Liste umgewandelt, dann wird die L√§nge der Lsit berechnet.

import ast
def get_list_len(emojies):
    l = ast.literal_eval(emojies)
    return len(l)

df_tweetskb_emojies['Anzahl der Emojis'] = df_tweetskb_emojies.apply(lambda row: get_list_len(row['Emojie']), axis=1)
print('Anzahl der Emojis wurde berechnet!')
```

```python
#Hier wird gepr√ºft, ob die Anzahl der Emojis berechnet wurde. 
df_tweetskb_emojies[df_tweetskb_emojies['Anzahl der Emojis']>0].head()
```

```python
#Hier wird gepr√ºft, ob es Null-Daten in datframe gibt.
df_tweetskb_emojies['Anzahl der Emojis'].isna().sum()
```

```python
#In dieser zwei Methoden wird zuerst die String-Daten zu Liste umgewandelt.
#F√ºr jedes Einzelne Emoji wird die Stimmung bewertet. Vier Werte, die f√ºr die Stimmung berechnet wurden, werden gespeichert.
#Nicht alle Emojis werden bewertet. Daher werden die berechneten Werte aufsummiert und durch die Anzahl der Emojis, die Bewertet wurden, geteilt.
#Die Anzahl der Emojis die bewertet wurden, werden auch gespeichert.
#Alle Ergebnisse werden mit einem Leerzeichen getrennt und gespeichert.

def get_emoji_l(text):
    return ast.literal_eval(text)

    
def get_emoji_bewertung(l):
    list_emo = get_emoji_l(l)
    zaehler = 0.0
    neg = 0.0
    neu = 0.0
    pos = 0.0
    sent = 0.0
    s = ''
    laenge = len(list_emo)
    for i in range(0, laenge):
        dic = get_emoji_sentiment_rank(list_emo[i])
        #print(dic)
        if dic != 0.0:
            if (dic['sentiment_score'] != 1 and dic['sentiment_score'] != -1) :
                zaehler += 1

                neg = neg + dic['negative']
                neu = neu + dic['neutral']
                pos = pos + dic['positive']
                sent = sent + dic['sentiment_score']

                summe = neg + neu +pos

                neg = float( "{:.2f}".format(round((neg / summe), 2)) )
                neu = float( "{:.2f}".format(round((neu / summe), 2)) )
                pos = float( "{:.2f}".format(round((pos / summe), 2)) )
            
        else:
            neg = neg + 0.0
            neu = neu + 0.0
            pos = pos + 0.0
    if (laenge == 0.0) or ( zaehler == 0.0) :
        s = s + str(neg) + ' ' + str(neu) + ' ' + str(pos) + ' ' + str(sent)+ ' ' + str(zaehler)
    else :
        s = s + str(neg /zaehler ) + ' ' + str(neu / zaehler) + ' ' + str(pos / zaehler) + ' ' + str(sent / zaehler)+ ' ' + str(zaehler)

    return s
```

```python
import ast

df_tweetskb_emojies['Emojis_Bewertung'] = df_tweetskb_emojies.apply(lambda row: get_emoji_bewertung(row['Emojie']), axis=1)
print('Emojies wurden bewertet!')
```

```python
#Die Methode teilt die String-Daten, multipliziert ersten Teil mit -1 und gibt ihn zur√ºck.
def get_neg(emoji):
    s = emoji.split()
    return -1 * float(s[0])
```

```python
#Die Methode teilt die String-Daten, und gibt den zweiten Teil zur√ºck.
def get_neu(emoji):
    s = emoji.split()
    return float(s[1])
```

```python
#Die Methode teilt die String-Daten, und gibt den dritten Teil zur√ºck.
def get_pos(emoji):
    s = emoji.split()
    return float(s[2])
```

```python
#Die Methode teilt die String-Daten, und gibt den vierten Teil zur√ºck.
def get_compound(emoji):
    s = emoji.split()
    return (float(s[3]))
```

```python
#Die Methode teilt die String-Daten, und gibt den f√ºnften Teil zur√ºck.
def get_b_zahl(emoji):
    s = emoji.split()
    return float(s[4])
```

```python
#Hier wird die berechnete negative Stimmung f√ºr die Emojis getrennt gespeichert 
df_tweetskb_emojies['Emoji_neg'] = df_tweetskb_emojies.apply(lambda row: get_neg(row['Emojis_Bewertung']) , axis=1)
print('Emoji neg!')
```

```python
#Hier wird die berechnete neutrale Stimmung f√ºr die Emojis getrennt gespeichert 
df_tweetskb_emojies['Emoji_neu'] = df_tweetskb_emojies.apply(lambda row: get_neu(row['Emojis_Bewertung']) , axis=1)
print('Emoji neu!')
```

```python
#Hier wird die berechnete positive Stimmung f√ºr die Emojis getrennt gespeichert 
df_tweetskb_emojies['Emoji_pos'] = df_tweetskb_emojies.apply(lambda row: get_pos(row['Emojis_Bewertung']) , axis=1)
print('Emoji pos!')
```

```python
#Hier wird den berechneten Mittelwert f√ºr die Emojis getrennt gespeichert 
df_tweetskb_emojies['Emoji Stimmung Mittelwert'] = df_tweetskb_emojies.apply(lambda row: get_compound(row['Emojis_Bewertung']) , axis=1)
print('Emoji sent!')
```

```python
#Hier wird die Anzahl der bewerteten Emojis gespeichert 
df_tweetskb_emojies['Bewertete Emojis'] = df_tweetskb_emojies.apply(lambda row: get_b_zahl(row['Emojis_Bewertung']) , axis=1)
print('Bewertete Emojis!')
```

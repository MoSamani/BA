# Benutzer Features(Ohne Polarisierungen)
#### Zuerst werden die Tweets in TweetsCov19 bearbeitet, dann die Tweets in TweetsKB

- Anzahl der geposteten Tweets wird berechnet
- Anzahl der gelöschten Tweest wird berechnet
- Löschungsquote wird berechnet
- Reputaion wird berechnet

***

**TweetsCov19:**

```python
#Labels und Pfad für den TweetsCov19
TweetsCov19_labels = ['Tweet_Id', "Benutzername", 'Zeitstempel', 'Abonnenten', 'Abonnierten', 'Retweets', 'Likes', 'Entities', 'Stimmung', 'Erwähnungen', 'Hashtags', 'URLs', 'Text', 'Ist_gelöscht', 'Anzahl der Erwähnungen', 'Anzahl der Hashtags', 'Anzahl der URLs', 'Domäne', 'Textlänge', 'Lesbarkeit']
TweetsCov19_Pfad =  'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/tweetscov19_sample/TweetsCov19_stichprobe_lesbarkeit.tsv'
```

```python
#Stichprobe von Tweetscov19 lesen
df_tweetscov19 = pd.read_csv(TweetsCov19_Pfad, sep='\t', error_bad_lines=False, names=TweetsCov19_labels)
print('Form:', df_tweetscov19.shape)
df_tweetscov19.head()
```

```python
#Dataframe einschränken
df_tweetscov19 = df_tweetscov19[['Tweet_Id', "Benutzername", 'Abonnenten', 'Abonnierten', 'Ist_gelöscht']]
```

```python
#Für alle gepostete Tweets eine 1 eintragen.
df_tweetscov19['num_posted'] = df_tweetscov19.apply(lambda row: 1, axis=1)
```

```python
#Für die gelöschten Tweets eine 1 eintragen
df_tweetscov19['num_del'] = df_tweetscov19.apply(lambda row: 1 if (row['Ist_gelöscht'] == True) else 0, axis=1)
```

```python
#Die geposteten und gelöschten Tweets nach Benutzername gruppieren und die Werte summieren
df_tweetscov19_loe_repu = df_tweetscov19
df_tweetscov19_unique_user = df_tweetscov19_loe_repu.groupby('Benutzername')['num_posted', 'num_del'].sum()
```

```python
#Anzahl der gelöschten Tweets durch die geposteten berechnen.
df_tweetscov19_unique_user['Deletion_ratio'] = df_tweetscov19_unique_user.apply(lambda row: row['num_del'] / row['num_posted'] , axis=1)
df_tweetscov19_unique_user.head()
```

```python
df_tweetscov19_unique_user.shape
```

```python
#Die Einzigartige Benutzer auswählen.
df_tweetscov19_loe_repu = df_tweetscov19_loe_repu.drop_duplicates(subset=['Username'])
df_tweetscov19_loe_repu.shape
```

```python
# Die Reputation für die Benutzer berechnen.
df_tweetscov19_loe_repu['Reputation'] = df_tweetscov19_loe_repu.apply(lambda row: row['Followers'] / (row['Followers'] + row['Friends']) if (row['Followers'] + row['Friends']) != 0 else 0, axis=1)
```

```python
#Die dataframe neu indexieren.
df_tweetscov19_unique_user = df_tweetscov19_unique_user.reset_index()
df_tweetscov19_unique_user.head()
```

```python
df_tweetscov19_loe_repu.head()
```

```python
#Die Reputation zu der Dataframe hinzufügen.
df = df_tweetscov19_loe_repu[['Username', 'Reputation']]
df_tweetscov19_posted_del = df_tweetscov19_posted_del[['Tweet_Id', 'Username', 'Followers', 'Friends', 'is_deleted',]]
df_tweetscov19_user = pd.merge(df_tweetscov19_posted_del, df, on='Username', how='inner')
print('Shape :', df_tweetscov19_user.shape)
df_tweetscov19_user.head()
```

```python
#Anzahl der geposteten, gelöschten und Löschungsquote zu der Dataframe hinzufügen.
df = df_tweetscov19_unique_user
df_tweetscov19_user = pd.merge(df_tweetscov19_user, df, on='Username', how='inner')
print('Shape :', df_tweetscov19_user.shape)
df_tweetscov19_user.head()
```

```python
#Daten speichern.
pfad_tsv3 = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/TweetsCov19_sample/TweetsCov19_alle_geloeschten/TweetsCov19_all_repu_del_.tsv'

with open(pfad_tsv3, 'w', newline='', encoding="utf-8") as out_file:
    tsv_writer = csv.writer(out_file, delimiter="\t")
    for idx, row in df_tweetscov19_user.iterrows():
        tsv_writer.writerow(row.values.tolist())
```

**TweetsKB:**

- Anzahl der geposteten Tweets berechnen
- Anzahl der gelöschten Tweets berechnen
- Löschungsquote berechnen
- Reputaion berechnen

***

```python
#Dataframe TweetsKB-Stichprobe lesen.
pfad_tsv = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/tweetskb_sample/tweetskb_sample_emojie_2.tsv'

df_tweetskb = pd.read_csv(pfad_tsv, sep='\t', error_bad_lines=False, names=['Tweet_Id', "Username", 'Timestamp', 'Followers', 'Friends', 'Retweets', 'Favorites', 'Text', 'Urls', 'is_deleted', 'Hashtags', 'Mentions', 'vader_pos', 'vader_neu', 'vader_neg', 'vader_com', 'Hashtags_num', 'Mentions_num', 'Text_len', 'URLs_dom', 'URLs_num', 'readability', 'Emojie', 'Emoji_num','Emoji_neg', 'Emoji_neu', 'Emoji_pos', 'Emoji_sent', 'Emoji_b_zahl' ])
pd.options.display.max_columns = None
print('Shape :', df_tweetskb.shape)
df_tweetskb.head()
```

```python
#Dataframe wird eingeschränkt.
# Für die geposteten Tweets wird eine 1 eingetragen.
#Für alle gelöschten Tweets wird ein 1 getragen.
df_tweetskb_posted_del = df_tweetskb[['Tweet_Id', "Username", 'Followers', 'Friends', 'is_deleted']]
df_tweetskb_posted_del['num_posted'] = df_tweetskb_posted_del.apply(lambda row: 1, axis=1)
df_tweetskb_posted_del['num_del'] = df_tweetskb_posted_del.apply(lambda row: 1 if (row['is_deleted'] == True) else 0, axis=1)
df_tweetskb_posted_del.head()
```

```python
#Die Summer der Anzahl der geposteten und gelöschten tweets wird berehnet.
df_tweetskb_unique_user = df_tweetskb_posted_del.groupby('Username')['num_posted', 'num_del'].sum()
```

```python
#Anzahl der gelöschten durch die geposteten wird berechnet
df_tweetskb_unique_user['Deletion_ratio'] = df_tweetskb_unique_user.apply(lambda row: row['num_del'] / row['num_posted'] , axis=1)
```

```python
#Die Reputation wird berechnet.
df_tweetskb_posted_del = df_tweetskb_posted_del.drop_duplicates(subset=['Username'])
print(df_tweetskb_posted_del.shape)
df_tweetskb_posted_del['Reputation'] = df_tweetskb_posted_del.apply(lambda row: row['Followers'] / (row['Followers'] + row['Friends']) if (row['Followers'] + row['Friends']) != 0 else 0, axis=1)
df_tweetskb_posted_del.head()
```

```python
#Dataframe wird neu indexieret.
df_tweetskb_unique_user = df_tweetskb_unique_user.reset_index()
df_tweetskb_unique_user.head()
```

```python
#Anzahl der geposteten und gelöschten Tweets, sowie die Löschungsquote wird zu Dder ataframe hinzugefügt.
df = df_tweetskb_unique_user[['Username', 'num_posted', 'num_del', 'Deletion_ratio']]
df_tweetskb = df_tweetskb[['Tweet_Id', "Username", 'Followers', 'Friends', 'is_deleted']]
df_tweetskb_user = pd.merge(df_tweetskb, df, on='Username', how='inner')
print('Shape :', df_tweetskb_user.shape)
df_tweetskb_user.head()
```

```python
#Reputation wird berechnet.
df = df_tweetskb_posted_del[['Username', 'Reputation']]
df_tweetskb_user = pd.merge(df_tweetskb_user, df, on='Username', how='inner')
print('Shape :', df_tweetskb_user.shape)
df_tweetskb_user.head()
```

```python
#Dataframe wird gespeichert.
pfad_tsv3 = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/Tweetskb_sample/TweetsKB_alle_geloeschten/TweetsKB_all_repu_del_.tsv'

with open(pfad_tsv3, 'w', newline='', encoding="utf-8") as out_file:
    tsv_writer = csv.writer(out_file, delimiter="\t")
    for idx, row in df_tweetskb_user.iterrows():
        tsv_writer.writerow(row.values.tolist())
```

```python
#CSV Bibliothek importieren 
import csv
#Pandas importieren und als pd bezeichnen
import pandas as pd
#Numpy importieren und als np bezeichnen 
import numpy as np
#Json Biliothek importieren
import json
#Obere Schranke für die Größe der Daten auf 100,000,000 setzen  
csv.field_size_limit(100000000)
#Maximale Anzahl von Zeilen und Spalten Zeigen 
pd.options.display.max_columns = None
pd.options.display.max_rows = None
#Um die Zeiträume bearbeiten zu können
from datetime import datetime
#Seaborn für die Plot importieren
import seaborn as sns
#matplotlib Importieren und als plt bezeichnen
import matplotlib.pyplot as plt
#Macht die Plots inline
%matplotlib inline 
#dates importieren, um Zeitstempel bearbeiten zu können
import matplotlib.dates as mdates
#tick importieren
import matplotlib.ticker as tick
#Die Nachkommastellen auf 4 Zashlen einschränken
pd.options.display.float_format = '{:.4f}'.format
```

# TweetsCov19

***

```python
#Die geposteten Tweets von Benutzer, die in allen drei Dimensionen einen polarisierten Wert haben. 
pfad_tsv = '../Data/DAtaframes/tweetscov19_time_user_1.tsv'

df_tweetscov19_user = pd.read_csv(pfad_tsv, sep='\t', error_bad_lines=False, names=['Tweet_Id', 'Username', 'is_deleted', 'Retweets', 'Favorites', 'vader_com', 'Hashtags_num', 'Mentions_num', 'Text_len', 'URLs_num', 'readability', 'Emoji_num', 'Emoji_sent', 'politic_pol', 'science_pol', 'moderacy_pol', 'porn_shalla', 'news_shalla', 'socialnet_shalla', 'social', 'family', 'friend', 'female', 'male' , 'work', 'leisure', 'home', 'money', 'relegion', 'death', 'informal', 'swear', 'netspeak', 'assent', 'nonflu', 'filler', 'body', 'health', 'sexual', 'ingest', 'posemo', 'negemo', 'anx', 'anger', 'sad', 'insight', 'cause', 'discrep', 'tentat', 'certain', 'differ', 'affiliation', 'achiev', 'power', 'reward', 'risk', 'focuspast', 'focuspresent', 'focusfuture', 'i', 'we', 'you', 'shehe', 'they', 'ipron', 'article', 'prep', 'auxverb', 'adverb', 'conj', 'negate', 'verb', 'adj', 'compare', 'interrog', 'number', 'quant', 'see', 'hear', 'feel', 'motion', 'space', 'time', 'politic', 'science', 'moderacy','posted','deleted', 'Deletion_ratio', 'Reputation' ])
pd.options.display.max_columns = None
print('Shape :', df_tweetscov19_user.shape)
df_tweetscov19_user.head()
```

```python
#Kerndichteschätzung der geposteten und gelöschten Tweets in TweetsCov19
#Englische Labels
def plot_kdeplot(df, hue, x, x_label, titel, name): 
    ax = sns.kdeplot(data=df, x=x, hue=hue, fill=True, legend= False)
      
    ax.set_ylabel(ylabel='Density', fontsize=15)
    ax.set_xlabel(xlabel=x_label, fontsize=17)

    ax.tick_params(axis='x', which='major', labelsize=16)
    plt.legend(labels = ['Deleted','Non-deleted'], bbox_to_anchor=(0.2, 1.13), ncol=2, loc=2, borderaxespad=0, fontsize=12)

    plt.tight_layout()
    plt.savefig('./plots_pdfs/'+name+'.pdf')
```

```python
#Kerndichteschätzung der geposteten und gelöschten Tweets in Bezug auf Politische Dimension
df = df_tweetscov19_user[['is_deleted', 'politic']]
plot_kdeplot(df, 'is_deleted','politic' ,'Political' ,'TweetsCov19', 'kdeplot_tweetsocv19_politic_schnitt')
```

```python
#Kerndichteschätzung der geposteten und gelöschten Tweets in Bezug auf wissenschaftliche Dimension Dimension
df = df_tweetscov19_user[['is_deleted', 'science']]
plot_kdeplot(df, 'is_deleted', 'science', 'Science', 'TweetsCov19', 'kdeplot_tweetsocv19_science_schnitt')
```

```python
#Kerndichteschätzung der geposteten und gelöschten Tweets in Bezug auf moderate Dimension
df = df_tweetscov19_user[['is_deleted', 'moderacy']]
plot_kdeplot(df, 'is_deleted', 'moderacy','Moderacy' ,'TweetsCov19', 'kdeplot_tweetsocv19_moderacy_schnitt')
```

```python
#Kerndichteschätzung der geposteten und gelöschten Tweets in TweetsCov19
def plot_kdeplot_de(df, hue, x, x_label, titel, name): 
    ax = sns.kdeplot(data=df, x=x, hue=hue, fill=True, legend= False)
      
    ax.set_ylabel(ylabel='Dichte', fontsize=15)
    ax.set_xlabel(xlabel=x_label, fontsize=17)

    ax.tick_params(axis='x', which='major', labelsize=16)
    plt.legend(labels = ['Gelöscht','Nicht gelöscht'], bbox_to_anchor=(0.18, 1.13), ncol=2, loc=2, borderaxespad=0, fontsize=12)

    plt.tight_layout()
    plt.savefig('plots_pdfs/'+name+'.pdf')
```

```python
#Kerndichteschätzung der geposteten und gelöschten Tweets in Bezug auf politische Dimension
df = df_tweetscov19_user[['is_deleted', 'politic']]
plot_kdeplot_de(df, 'is_deleted','politic' ,'Politische' ,'TweetsCov19', 'kdeplot_tweetsocv19_politic_schnitt')
```

```python
#Kerndichteschätzung der geposteten und gelöschten Tweets in Bezug auf wissenschaftliche Dimension
df = df_tweetscov19_user[['is_deleted', 'science']]
plot_kdeplot_de(df, 'is_deleted', 'science', 'Wissenschaftliche', 'TweetsCov19', 'kdeplot_tweetsocv19_science_schnitt')
```

```python
#Kerndichteschätzung der geposteten und gelöschten Tweets in Bezug auf moderate Dimension
df = df_tweetscov19_user[['is_deleted', 'moderacy']]
plot_kdeplot_de(df, 'is_deleted', 'moderacy','Moderate' ,'TweetsCov19', 'kdeplot_tweetsocv19_moderacy_schnitt')
```

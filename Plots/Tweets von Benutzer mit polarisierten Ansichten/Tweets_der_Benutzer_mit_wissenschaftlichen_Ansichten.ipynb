```python
#CSV Bibliothek importieren 
import csv
#Pandas importieren und als pd bezeichnen
import pandas as pd
#Numpy importieren und als np bezeichnen 
import numpy as np
#Json Biliothek importieren
import json
#Obere Schranke für die Größe der Daten auf 100,000,000 setzen  
csv.field_size_limit(100000000)
#Maximale Anzahl von Zeilen und Spalten Zeigen 
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

# Verfasste Tweets von Befürwörter und Gegner wissenschaftlichen Meinungen

#### Zuerst werden die Tweets der Benutzer in TweetsCov19 bearbeitet, dann die Tweets der Benutzer in TweetsKB

1. Die geposteten Tweets von anerkannten Benutzer mit polarisierten wissenschaftlichen Ansichten werden aussortiert.
2. Die Anzahl der geposteten und gelöschten Tweets, Löschungsquote und Reputation wird zu Dataframe hinzugefügt.
3. Die Feature Ist_gelöscht wird zu dataframe hinzugefügt, damit die gelöschten Tweets zu finden wären.
4. Dataframe wird gespeichert.

***

### TweetsCov19:

```python
#Die Dataframe, die Informationen über die Domäne enthält und sich auf Benutzer mit bekannten wissenschaftlichen Meinungen bezieht, wird gelesen.
pfad_tsv1 = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/dom_label/tweetscov19_unique_user_dom_science.tsv'

df_tweetscov19_science = pd.read_csv(pfad_tsv1, sep='\t', error_bad_lines=False, names=['Username', 'Urls_num', 'neg', 'pos', 'pol_num', 'science'])
pd.options.display.max_columns = None
print('Shape :', df_tweetscov19_science.shape)
df_tweetscov19_science.head()
```

```python
#Die Dataframe, die die Anzhal der geposteten und gelöschten Tweets sowie Löschungsquote und Reputation enthält wirde hier gelesen.
pfad_tsv = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/TweetsCov19_sample/TweetsCov19_alle_geloeschten/TweetsCov19_all_repu_del_.tsv'

df_tweetscov19_user_repu_del = pd.read_csv(pfad_tsv, sep='\t', error_bad_lines=False, names=['Tweet_Id', 'Username', 'Followers', 'Friends', 'is_deleted', 'Reputation', 'num_posted', 'num_del', 'Deletion_ratio'])
pd.options.display.max_columns = None
print('Form :', df_tweetscov19_user_repu_del.shape)
df_tweetscov19_user_repu_del.head()
```

```python
#Die Anzhal der geposteten und gelöschten Tweets sowie Löschungsquote und Reputation wird zu der Benutzer mit bekannten pro- oder anti-wissenschaftlichen Ansichten hinzugefügt.
df_tweetscov19_user_repu_del = df_tweetscov19_user_repu_del[['Tweet_Id', 'Username', 'num_posted', 'num_del', 'Deletion_ratio', 'Reputation']]
df_tweetscov19_science = pd.merge( df_tweetscov19_user_repu_del, df_tweetscov19_science, on='Username', how='inner')
print('Form :', df_tweetscov19_science.shape)
df_tweetscov19_science.head()
```

```python
#Die TweetsCov19-Stichprobe wird gelesen.
pfad_tsv1 = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/TweetsCov19_sample/tweetsCov19_sample_2.tsv'

df_tweetscov19 = pd.read_csv(pfad_tsv1, sep='\t', error_bad_lines=False, names=['Tweet_Id', 'Username', 'Timestamp', 'Followers', 'Friends', 'Retweets', 'Favorites', 'Entities', 'Sentiment', 'Mentions', 'Hashtags', 'URLs', 'text', 'is_deleted', 'Mentions_num','Hashtags_num', 'Urls_num', 'Url_dom', 'Vader_neg', 'Vader_neu', 'Vader_pos', 'Vader_com', 'Sentiment_pos', 'Sentiment_neg', 'Sentiment_sum', 'fk_readability', 'fk_grade_level', 'Text_len', "Entities_len", 'Emojies', 'Emoji_num',  'Emoji_neg', 'Emoji_neu', 'Emoji_pos', 'Emoji_sentiment', 'Emoji_bearbeiteten'])
pd.options.display.max_columns = None
print('Shape :', df_tweetscov19.shape)
df_tweetscov19.head()
```

```python
#Die Feature 'is_deleted' wird von TweetsCov19 ausgewählt und zu der Tweets hinzugefügt.
#Dadurch wird die Information zu dem Tweet hinzugefügt, ob der Tweet gelöscht wurde oder nicht.
df_tweetscov19 = df_tweetscov19[['Tweet_Id', 'is_deleted']]
df_tweetscov19_science = pd.merge(df_tweetscov19_science, df_tweetscov19, on='Tweet_Id', how='inner')
print('Shape :', df_tweetscov19_science.shape)
df_tweetscov19_science.head()
```

```python
#Die neue Dataframe wird gespeichert.
pfad_tsv3 = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/TweetsCov19_sample/df_polarisierungen/tweetscov19_user_tweets_science_all.tsv'

with open(pfad_tsv3, 'w', newline='', encoding="utf-8") as out_file:
    tsv_writer = csv.writer(out_file, delimiter="\t")
    for idx, row in df_tweetscov19_science.iterrows():
        tsv_writer.writerow(row.values.tolist())
```

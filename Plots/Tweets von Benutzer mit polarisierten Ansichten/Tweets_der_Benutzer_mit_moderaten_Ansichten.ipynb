# Verfasste Tweets von Moderaten und Hardliner

#### Zuerst werden die Tweets der Benutzer in TweetsCov19 bearbeitet, dann die Tweets der Benutzer in TweetsKB

1. Die geposteten Tweets von anerkannten Benutzer mit polarisierten moderaten Ansichten werden ausgew√§hlt.
2. Die Anzahl der geposteten und gel√∂schten Tweets, L√∂schungsquote und Reputation wird zu Dataframe hinzugef√ºgt.
3. Die Feature Ist_gel√∂scht wird zu dataframe hinzugef√ºgt, damit die gel√∂schten Tweets zu finden w√§ren.
4. Dataframe wird gespeichert.

***

### TweetsCov19:

```python
#Die Dataframe, die Informationen √ºber die Dom√§ne enth√§lt und sich auf Benutzer mit bekannten moderaten Meinungen bezieht, wird gelesen.
pfad_tsv1 = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/dom_label/tweetscov19_unique_user_dom_moderacy.tsv'

df_tweetscov19_moderacy = pd.read_csv(pfad_tsv1, sep='\t', error_bad_lines=False, names=['Username', 'Urls_num', 'neg', 'pos', 'Url_dom', 'moderacy'])
pd.options.display.max_columns = None
print('Shape :', df_tweetscov19_moderacy.shape)
df_tweetscov19_moderacy.head()
```

```python
#Die Dataframe, die die Anzhal der geposteten und gel√∂schten Tweets sowie L√∂schungsquote und Reputation enth√§lt wirde hier gelesen.
pfad_tsv = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/TweetsCov19_sample/TweetsCov19_alle_geloeschten/TweetsCov19_all_repu_del_.tsv'

df_tweetscov19_user_repu_del = pd.read_csv(pfad_tsv, sep='\t', error_bad_lines=False, names=['Tweet_Id', 'Username', 'Followers', 'Friends', 'is_deleted', 'Reputation', 'num_posted', 'num_del', 'Deletion_ratio'])
pd.options.display.max_columns = None
print('Form :', df_tweetscov19_user_repu_del.shape)
df_tweetscov19_user_repu_del.head()
```

```python
#Die Anzhal der geposteten und gel√∂schten Tweets sowie L√∂schungsquote und Reputation wird zu der Benutzer mit bekannten moderaten Ansichten hinzugef√ºgt.
df_tweetscov19_user_repu_del = df_tweetscov19_user_repu_del[['Tweet_Id', 'Username', 'num_posted', 'num_del', 'Deletion_ratio', 'Reputation']]
df_tweetscov19_moderacy = pd.merge( df_tweetscov19_user_repu_del, df_tweetscov19_moderacy, on='Username', how='inner')
print('Form :', df_tweetscov19_moderacy.shape)
df_tweetscov19_moderacy.head()
```

```python
#Die TweetsCov19-Stichprobe wird gelesen.
pfad_tsv1 = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/TweetsCov19_sample/tweetsCov19_sample_2.tsv'

df_tweetscov19 = pd.read_csv(pfad_tsv1, sep='\t', error_bad_lines=False, names=['Tweet_Id', 'Username', 'Timestamp', 'Followers', 'Friends', 'Retweets', 'Favorites', 'Entities', 'Sentiment', 'Mentions', 'Hashtags', 'URLs', 'text', 'is_deleted', 'Mentions_num','Hashtags_num', 'Urls_num', 'Url_dom', 'Vader_neg', 'Vader_neu', 'Vader_pos', 'Vader_com', 'Sentiment_pos', 'Sentiment_neg', 'Sentiment_sum', 'fk_readability', 'fk_grade_level', 'Text_len', "Entities_len", 'Emojies', 'Emoji_num',  'Emoji_neg', 'Emoji_neu', 'Emoji_pos', 'Emoji_sentiment', 'Emoji_bearbeiteten'])
pd.options.display.max_columns = None
print('Shape :', df_tweetscov19.shape)
df_tweetscov19.head()
```

```python
#Die Feature 'is_deleted' wird von TweetsCov19 ausgew√§hlt und mit der aus moderaten Benutzer bestehenden Dataframe vereinigt.
#Dadurch wird die Information zu dem Tweet hinzugef√ºgt, ob der Tweet gel√∂scht wurde oder nicht.
df_tweetscov19 = df_tweetscov19[['Tweet_Id', 'is_deleted']]
df_tweetscov19_moderacy = pd.merge(df_tweetscov19_moderacy, df_tweetscov19, on='Tweet_Id', how='inner')
print('Shape :', df_tweetscov19_moderacy.shape)
df_tweetscov19_moderacy.head()
```

```python
#Die neue Dataframe wird gespeichert.
pfad_tsv3 = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/TweetsCov19_sample/df_polarisierungen/tweetscov19_user_tweets_moderacy_all.tsv'

with open(pfad_tsv3, 'w', newline='', encoding="utf-8") as out_file:
    tsv_writer = csv.writer(out_file, delimiter="\t")
    for idx, row in df_tweetscov19_moderacy.iterrows():
        tsv_writer.writerow(row.values.tolist())
```

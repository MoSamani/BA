# TweetsCov19
#### In diesem Notebook werden folgende Schritte unternommen:
1. Die Inhalte der hydrierten Tweets werden in Dataframe eingetragen
2. Die gelöschten Tweets werden mithilfe der heydrierten Tweets identifiziert

```python
#Diese Methode list jede Zeile von Jsonline, nimmt die TWeetID und den Text von jedem Tweet, und speichert sie. 

def id_Text_von_jsonl_in_tsv(pfad_jsonl, pfad_tsv):
    zaehler = 0
    with open(pfad_jsonl,'r',encoding="utf8") as jl:
        with open (pfad_tsv, 'w' , newline='', encoding="utf8") as hyd:
            writer = csv.writer(hyd, delimiter='\t')
            for t in jl:
                tweet = json.loads(t)
                #print(tweet)
                l = []
                l.append(int(tweet['id']))
                l.append(str(tweet['full_text']))
                #print(l)
                writer.writerow(l)
                
                zaehler += 1
    print('zaehler: ', zaehler)
```

```python
pfad1 = 'C:/Mo/tweetDB/tweetscov19/tweetIDs/TweetsCOV19_hydrated.jsonl'
pfad2 = 'C:/Mo/tweetDB/tweetscov19/tweetIDs/TweetsCOV19_hydrated_id_text.tsv'

#Die Funktion wird gerufen 
id_Text_von_jsonl_in_tsv(pfad1, pfad2)
```

```python
lables = ['Tweet_Id', "Benutzername", 'Zeitstempel', 'Abonnenten', 'Abonnierten', 'Retweets', 'Likes', 'Entities', 'Stimmung', 'Erwähnungen', 'Hashtags', 'URLs']
pfad_tweetscov19_1 = 'C:/Mo/tweetDB/tweetscov19/tweetIDs/TweetsCOV19.tsv'
pfad_tweetscov19_2 = 'C:/Mo/tweetDB/tweetscov19/tweetIDs/TweetsCOV19_052020.tsv'
pfad_tweetscov19_3 = 'C:/Mo/tweetDB/tweetscov19/tweetIDs/TweetsCOV19_3.tsv'

#TweetsCov19 besteht aus drei Teilen
# Hier wird erster Teil von tweetsCov19 gelesen

tweetsCov19_1 = pd.read_csv(pfad_tweetscov19_1, sep='\t', error_bad_lines=False, names=lables)
print('Form:', tweetsCov19_1.shape )
tweetsCov19_1.head()
```

```python
# Hier wird zweiter Teil von tweetsCov19 gelesen

tweetsCov19_2 = pd.read_csv(pfad_tweetscov19_2, sep='\t', error_bad_lines=False, names=lables)
print('Form:', tweetsCov19_2.shape )
tweetsCov19_2.head()
```

```python
# Hier wird dritter Teil von tweetsCov19 gelesen

tweetsCov19_3 = pd.read_csv(pfad_tweetscov19_3, sep='\t', error_bad_lines=False, names=lables)
print('Form:', tweetsCov19_3.shape )
tweetsCov19_3.head()
```

```python
# Hier werden drei Teile von TweetsCov19 zueinander angehängt.

tweetsCov19_1 = tweetsCov19_1.append(tweetsCov19_2, ignore_index=True)
tweetsCov19 = tweetsCov19_1.append(tweetsCov19_3, ignore_index=True)
print('Die Form von TweetsCov19:', tweetsCov19.shape)
tweetsCov19.head()
```

```python
#Pfad von TweetsCov19-Dataframe
pfad_tweetscov19_hydrated_texts = 'C:/Mo/tweetDB/tweetscov19/tweetIDs/TweetsCOV19_hydrated_id_text.tsv'

# Die Ids und Texts von den Hydrieretn Tweets lesen

tweetscov19_hydrated_texts = pd.read_csv(pfad_tweetscov19_hydrated_texts, sep='\t', error_bad_lines=False, names=['Tweet_Id', 'text'])
pd.options.display.max_columns = None
print('Shape:', tweetscov19_hydrated_texts.shape )
tweetscov19_hydrated_texts.head()
```

```python
#Hier werden die Texts von den hydrierten tweets zu Dataframe hinzugefügt.
tweetsCov19 = pd.merge(tweetsCov19, tweetscov19_hydrated_texts, on='Tweet_Id', how='left')
print('Form von TweetsCov19:', tweetsCov19.shape)
```

```python
#Hier wir eine neue Spalte für Dataframe erzeugt.
#Wenn ein Text in der Zeile existiert, wird die Feature 'Ist_gelöscht'  mit True, und es kein Text existiert wird es mit False markiert. 

tweetsCov19['Ist_gelöscht'] = tweetsCov19.apply(lambda row: True if pd.isnull(row['text']) else False, axis=1)
tweetsCov19.head()
```

```python
#In diesem Teil werden die Ids von gelöschten Tweets aussortiert und gespeichert

df_gelöscht = tweetsCov19[tweetsCov19['Ist_gelöscht'] == True]
df_gelöscht = df_gelöscht[['Tweet_Id']]

pfad_ids_gelöscht = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/tweetscov19_gelöscht_ids.tsv'

with open(pfad_ids_gelöscht, 'w', newline='', encoding="utf-8") as out_file:
        tsv_writer = csv.writer(out_file, delimiter="\t")
        for idx, row in df_gelöscht.iterrows():
            tsv_writer.writerow(row.values.tolist())
```

```python
#Labels von Features der gelsöchten Tweets
labales_gelöschte_Tweets = ['Tweet_Id', "Benutzername", 'Zeitstempel', 'Abonnenten', 'Abonnierten', 'Retweets', 'Likes', 'text_gelöscht', 'URLs']
#Der Pfad zu gelöschten Tweets
pfad_tweetscov19_gelöscht = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/deleted_tweetscov19_tweets.tsv'

#Hier werden die gelöschten Tweets von TweetsCov19, die von dem TweetsKB-Repository heruntergeladen wurden, gelesen.
tweetsCov19_gelöscht = pd.read_csv(pfad_tweetscov19_gelöscht, sep='\t', error_bad_lines=False, names=labales_gelöschte_Tweets)
pd.options.display.max_columns = None
print('Shape:', tweetsCov19_gelöscht.shape )
tweetsCov19_gelöscht.head()
```

```python
#Die Texts werden hier von den hydrierten Tweets zu Dataframe hinzugefügt

tweetsCov19_gelöscht = tweetsCov19_gelöscht[['Tweet_Id', 'text_gelöscht']]
tweetsCov19 = pd.merge(tweetsCov19, tweetsCov19_gelöscht, on='Tweet_Id', how='left')
print('Form von TweetsCov19', tweetsCov19.shape)
tweetsCov19.head()
```

```python
#In diesem Teil wird eine Spalte, die Tweetinhalte von gelöschten und nicht gelöschten Tweets enthält erzeugt.
tweetsCov19['text'] = tweetsCov19.apply(lambda row: row['text_gelöscht'] if pd.isnull(row['text']) else row['text'], axis=1)
tweetsCov19.head(10)
```

```python
#Zuerst wird hier die Spalte text_gelöscht weggenommen
#Dann werden die Tweets nach ihrem Zeitstempel aufsteigend sortiert. 
tweetsCov19 = tweetsCov19[['Tweet_Id', "Benutzername", 'Zeitstempel', 'Abonnenten', 'Abonnierten', 'Retweets', 'Likes', 'Entities', 'Stimmung', 'Erwähnungen', 'Hashtags', 'URLs', 'text', 'Ist_gelöscht']]

tweetsCov19.Zeitstempel = pd.to_datetime(tweetsCov19.Zeitstempel)
tweetsCov19.sort_values(by = ['Zeitstempel'], inplace=True, ascending = True)
```

```python
start_datum = tweetsCov19.Zeitstempel.iloc[0]
end_datum = tweetsCov19.Zeitstempel.iloc[-1]
```

```python
print('Start : ', start_datum)
print('End   : ', end_datum)
```

```python
#Die im Zeitraum von 30.09.2019 bis 31.05.2020 verfassten Tweets werden hier aussortiert

start_datum = '2019-09-30 22:00:00+00:00'
end_datum  = '2020-05-31 22:59:54+00:00'

cond = (tweetsCov19['Zeitstempel'] > start_datum) & (tweetsCov19['Zeitstempel'] <= end_datum)
tweetsCov19 = tweetsCov19[cond]
print('Shape :', tweetsCov19.shape)
```

```python
#Hier wird eine Stichprobe von 9,409,840 Tweets von der Dataframe gezogen
tweetsCov19 = tweetsCov19.sample(n = 9409841)
print('Shape :', tweetsCov19.shape)
```

```python
#Da wird die Stichprobe gespeichert.
pfad_Stichprobe = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/tweetscov19_stichprobe.tsv'

with open(pfad_Stichprobe, 'w', newline='', encoding="utf-8") as out_file:
        tsv_writer = csv.writer(out_file, delimiter="\t")
        for idx, row in tweetsCov19.iterrows():
            tsv_writer.writerow(row.values.tolist())
```

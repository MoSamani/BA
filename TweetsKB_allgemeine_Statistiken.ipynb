```python
#Labels und Pfad

TweetsKB_labels    = ['Tweet_Id', 'Benutzername', 'Zeitstempel', 'Abonnenten', 'Abonnierten', 'Retweets', 'Likes', 'Text', 'URLs', 'Ist_gel√∂scht', 'Hashtags', 'Erw√§hnungen', 'Anzahl der Hashtags', 'Anzahl der Mentions', 'Textl√§nge', 'Lesbarkeit', 'Dom√§ne', 'Anzahl der URLs', 'vader pos', 'vader neu', 'vader neg', 'vader mittelwert' , 'Emojis', 'Anzahl der Emojis','Emoji_neg', 'Emoji_neu', 'Emoji_pos', 'Emoji_Mittelwert', 'Emoji_bearbeiteten']
TweetsKB_Pfad =  'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/tweetskb_sample/Tweetskb_stichprobe_final.tsv'
```

```python
#TweetsKB-Stichprobe wird gelesen

df_tweetskb = pd.read_csv(TweetsKB_Pfad, sep='\t', error_bad_lines=False, names=TweetsKB_labels)
print('Shape :', df_tweetskb.shape)
df_tweetskb.head()
```

```python
#Mittelwert, Median, Standardabweichung der Tweetsfeatures in gesammten Tweets 
df_tweetskb.describe()
```

```python
#Die gel√∂schten und nicht gel√∂schten Tweets werden getrennt.
tweetskb_del = df_tweetskb[df_tweetskb['Ist_gel√∂scht'] == True]
tweetskb_undel = df_tweetskb[df_tweetskb['Ist_gel√∂scht'] == False]
```

```python
#Mittelwert, Median, Standardabweichung der Tweetsfeatures in gel√∂schten Tweets 
tweetskb_del.describe()
```

```python
#Mittelwert, Median, Standardabweichung der Tweetsfeatures in nicht gel√∂schten Tweets 
tweetskb_undel.describe()
```

```python
print('Anzahl der Emojies in Tweetskb : ', df_tweetskb['Anzahl der Emojis'].sum())
print('Anzahl der Emojies in Tweetskb(gel√∂schte Tweets) : ', tweetskb_del['Anzahl der Emojis'].sum())
print('Anzahl der Emojies in Tweetskb(nicht gel√∂schte Tweets) : ', tweetskb_undel['Anzahl der Emojis'].sum())
```

```python
print('Anzahl der bearbeiteten Emojies in Tweetskb : ', df_tweetskb['Emoji_bearbeiteten'].sum())
```

```python
#Anzahl der Benutzer in TweetsKB
print('Anzahl der Benutezr : ', (df_tweetskb.Benutzername.unique().size))
```

```python
#Anzahl der gel√∂schten und nicht gel√∂schten Tweets
df_tweetskb.Ist_gel√∂scht.value_counts()
```

```python
#Anzahl der gel√∂schten Tweets und die Benutzer in gel√∂schten Tweets
print('Gel√∂schte Tweets :', tweetskb_del.shape)
print('Anzahl der Benutzer :', tweetskb_del.Benutzername.unique().size)
```

```python
#Anzahl der gel√∂schten Tweets und die Benutzer in nicht gel√∂schten Tweets
print('Nicht gel√∂schte Tweets :', tweetskb_undel.shape)
print('Anzahl der Benutzer :', tweetskb_undel.Benutzername.unique().size)
```

```python
#F√ºr alle geposteten Tweets wird 1 gespeichert.
df_tweetskb['num'] = df_tweetskb.apply(lambda row: 1, axis=1)
```

```python
#Die Tweets werden nach Benutzername gruppiert, und f√ºr jeden Tweet wird den Wert von 'num' um Eins erh√∂ht. 
df_by_username = df_tweetskb.groupby('Benutzername')['num'].count()
```

```python
#Mittelwert, Median, Standarddeviation f√ºr die Anzahl der geposteten Tweets in TweetsKB wird berechnet.
df_by_username.describe()
```

```python
#Die Tweets in gel√∂schten Tweets werden nach Benutzername gruppiert, und f√ºr jeden Tweet wird die Wertung von 'num' um Eins erh√∂ht. 
df_by_username_del = tweetskb_del.groupby('Benutzername')['num'].count()
```

```python
#Mittelwert, Median, Standardabweichung f√ºr die anzahl der gel√∂schten Tweets in gel√∂schten Tweets wird berechnet.
#Die Anzahl der gels√∂chten Tweets wird durch die Anzahl der Benutzer in gel√∂schten Tweets berechnet. 
df_by_username_del.describe()
```

```python
#Hier wird f√ºr die gel√∂schten Tweets jeweils eine Eins gespeichert.
df_tweetskb['num_del'] = df_tweetskb.apply(lambda row: 1 if (row['Ist_gel√∂scht'] == True) else 0, axis=1)
```

```python
#Die Tweets werden nach Benutzername gruppiert, und f√ºr jeden Tweet wird die Wwertung von 'num_del' um Eins erh√∂ht. 
df_by_username_del = df_tweetskb.groupby('Benutzername')['num_del'].sum()
```

```python
#Mittelwert, Median und Standardabweichung f√ºr die Anzahl der gel√∂schten Tweets in gesammten Tweeets wird berechnet.
#Die Anzahl der gels√∂chten Tweets wird durch die Anzahl der gesamten Benutzer in TweetsKB berechnet.
df_by_username_del.describe()
```

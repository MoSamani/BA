- Tweetspolarisierungungen bestimmen

```python
#Die Dataframe, die die Anzahl der benutzten liberalen und konservativen Domäne für jeden Tweets enthält, wird gelesen.   
pfad_tsv = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/dom_label/tweetsCov19_dom_politik.tsv'

df_politic = pd.read_csv(pfad_tsv, sep='\t', error_bad_lines=False, names=['Tweet_Id', 'Username','is_deleted','Urls_num', 'Url_dom', 'conservative', 'democrat'])
pd.options.display.max_columns = None
print('Shape :', df_politic.shape)
df_politic.head()
```

```python
#Für die politische Polarisierung der Tweets werden die Anzahl von Liberalen mit der Anzahl von Konservativen summiert.
#Jede liberale Domain bekommt eienen Wert von -1.
#Wenn das Ergebnis positiv wird, wird es als konservativ interpretiert. 
#Wenn das Ergebnis negativ wird, wird es als liberal interpretiert. 
df_politic['politic_pol'] = df_politic.apply(lambda row: (row['conservative'] + row['democrat']), axis=1)
```

```python
df_politic[df_politic['politic_pol']<0].head()
```

```python
#Die Dataframe, die die Anzahl der benutzten pro-wissenschaftlichen und anti-wissenschaftlichen Domäne für jeden Tweets enthält, wird gelesen.   
pfad_tsv = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/dom_label/tweetsCov19_dom_science.tsv'

df_science = pd.read_csv(pfad_tsv, sep='\t', error_bad_lines=False, names=['Tweet_Id', 'Username','is_deleted','Urls_num', 'Url_dom', 'Anti_science', 'pro_science'])
pd.options.display.max_columns = None
print('Shape :', df_science.shape)
df_science.head()
```

```python
#Für die wissenschaftliche Polarisierung der Tweets werden die Anzahl von pro-wissenschaftlichen Domäne mit der Anzahl von anti-wissenschaftlichen Domäne summiert.
#Jede Domain, die Anti-wissenschatliche Themen hat, bekommt eienen Wert von -1.
#Wenn das Ergebnis positiv wird, wird es als Befürwörter der wissenschaften Ansichten interpretiert. 
#Wenn das Ergebnis negativ wird, wird es als Gegner der Wissenschaft interpretiert. 
df_science['science_pol'] = df_science.apply(lambda row: (row['Anti_science'] + row['pro_science']), axis=1)
```

```python
df_science[df_science['science_pol']<0].head()
```

```python
#Die Dataframe, die die Anzahl der benutzten moderaten und hardliner Domäne für jeden Tweets enthält, wird hier gelesen.   
pfad_tsv = 'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/dom_label/tweetscov19_dom_moderacy.tsv'

df_moderacy = pd.read_csv(pfad_tsv, sep='\t', error_bad_lines=False, names=['Tweet_Id', 'Username','is_deleted','Urls_num', 'Url_dom', 'moderate', 'hardliner'])
pd.options.display.max_columns = None
print('Shape :', df_moderacy.shape)
df_moderacy.head()
```

```python
#Für die moderate Polarisierung der Tweets werden die Anzahl von Moderaten mit der Anzahl von Hardliner summiert.
#Jede moderate Domain bekommt eienen Wert von -1.
#Wenn das Ergebnis positiv wird, wird es als Moderat interpretiert. 
#Wenn das Ergebnis negativ wird, wird es als Hardliner interpretiert. 
df_moderacy['moderacy_pol'] = df_moderacy.apply(lambda row: (row['moderate'] + row['hardliner']), axis=1)
```

```python
df_moderacy[df_moderacy['moderacy_pol']<0].head()
```

```python
#Nur die Features TweetID und politische Polarisierung von Tweets werden ausgewählt.
df_politic = df_politic[['Tweet_Id','politic_pol']]
df_tweetscov19 = df_politic

print('Shape :', df_tweetscov19.shape)
df_tweetscov19.head()
```

```python
#Nur die Feature TweetID und wissenschaftliche Polarisierung von Tweets werden ausgewählt.
#Die wissenschaftliche Polarisierung der Tweets wird mit Dataframe bestehend aus politische Polarisierung vereinigt.
df_science = df_science[['Tweet_Id', 'science_pol']]
df_tweetscov19 = pd.merge(df_tweetscov19, df_science, on='Tweet_Id', how='left')
print('Shape :', df_tweetscov19.shape)
df_tweetscov19.head()
```

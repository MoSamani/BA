```python
#Labels und Pfad 
TweetsCov19_labels = ['Tweet_Id', "Benutzername", 'Zeitstempel', 'Abonnenten', 'Abonnierten', 'Retweets', 'Likes', 'Entities', 'Stimmung', 'Erwähnungen', 'Hashtags', 'URLs', 'Text', 'Ist_gelöscht', 'Anzahl der Erwähnungen', 'Anzahl der Hashtags', 'Anzahl der URLs', 'Domäne', 'Textlänge', 'Lesbarkeit','Vader_neg', 'Vader_neu', 'Vader_pos', 'Vader_Mittelwert', 'Emojis', 'Anzahl der Emojis',  'Emoji_neg', 'Emoji_neu', 'Emoji_pos', 'Emoji_Mittelwert', 'Emoji_bearbeiteten']
TweetsCov19_Pfad =  'C:/Mo/tweetDB/tweetscov19/TweetsCov19_tsv/data_sample/tweetscov19_sample/TweetsCov19_sample_original.tsv'
```

```python
#Stichprobe von Tweetscov19-Stichprobe lesen

df_tweetscov19 = pd.read_csv(TweetsCov19_Pfad, sep='\t', error_bad_lines=False, names=TweetsCov19_labels)
print('Form:', df_tweetscov19.shape)
df_tweetscov19.head()
```

```python
#Die gelöschten und nicht gelöschten Tweets werden hier getrennt.
df_tweetscov19_del = df_tweetscov19[df_tweetscov19['Ist_gelöscht'] == True]
df_tweetscov19_undel = df_tweetscov19[df_tweetscov19['Ist_gelöscht'] == False]
```

```python
#Anzahl der Benutzer in allen Tweets.
print('Anzahl der Benutezr : ', (df_tweetscov19.Benutzername.unique().size))
```

```python
#Anzahl der gelöschten und nicht gelöscten Tweets.
df_tweetscov19.Ist_gelöscht.value_counts()
```

```python
print('Gelöschte Tweets :', df_tweetscov19_del.shape)
print('Anzahl der Benutzer :', df_tweetscov19_del.Benutzername.unique().size)
```

```python
print('Nicht gelöschte Tweets :', df_tweetscov19_undel.shape)
print('Anzahl der Benutzer :', df_tweetscov19_undel.Benutzername.unique().size)
```

```python
#Mittelwert, Median, Standardabweichung der Tweetsfeatures in gesammten Tweets 
df_tweetscov19.describe()
```

```python
#Mittelwert, Median, Standardabweichung  der Tweetsfeatures in gelöschten Tweets 
df_tweetscov19_del.describe()
```

```python
#Mittelwert, Median, Standardabweichung der Tweetsfeatures in nicht gelöschten Tweets 
df_tweetscov19_undel.describe()
```

```python
print('Anzahl der Emojies in TweetsCov19 : ', df_tweetscov19['Anzahl der Emojis'].sum())
print('Anzahl der Emojies in TweetsCov19(gelöschte Tweets) : ', df_tweetscov19_del['Anzahl der Emojis'].sum())
print('Anzahl der Emojies in TweetsCov19(nicht gelöschte Tweets) : ', df_tweetscov19_undel['Anzahl der Emojis'].sum())
```

```python
print('Anzahl der bearbeiteten Emojies in TweetsCov19 : ', df_tweetscov19['Emoji_bearbeiteten'].sum())
```

```python
#Für alle geposteten Tweets wird 1 gespeichert.
df_tweetscov19['num'] = df_tweetscov19.apply(lambda row: 1, axis=1)
```

```python
#Die Tweets werden nach Benutzername gruppiert, und für jeden Tweet wird die Wertung von 'num' um Eins erhöht. 
df_by_username = df_tweetscov19.groupby('Benutzername')['num'].count()
```

```python
#Mittelwert, Median, Standarddeviation für die Anzahl der geposteten Tweets in TweetsCov19 wird berechnet.
df_by_username.describe()
```

```python
#Die Tweets in gelöschten Tweets werden nach Benutzername gruppiert, und für jeden Tweet wird die Wertung von 'num' um Eins erhöht. 
df_by_username_del = df_tweetscov19_del.groupby('Benutzername')['num'].count()
```

```python
#Mittelwert, Median, Standardabweichung für die Anzahl der gelöschten Tweets wird berechnet.
#Die Anzahl der gelsöchten Tweets wird durch die Anzahl der Benutzer in gelöschten Tweets berechnet. 
df_by_username_del.describe()
```

```python
#Hier wird für die gelöschten Tweets jeweils eine Eins gespeichert.
df_tweetscov19['num_del'] = df_tweetscov19.apply(lambda row: 1 if (row['Ist_gelöscht'] == True) else 0, axis=1)
```

```python
#Die Tweets werden nach Benutzername gruppiert, und für jeden Tweet wird den Wert von 'num_del' um Eins erhöht. 
df_by_username_del = df_tweetscov19.groupby('Benutzername')['num_del'].sum()
```

```python
#Mittelwert, Median und Standardabweichung für die Anzahl der gelöschten Tweets in gesammten Tweeets wird berechnet.
#Die Anzahl der gelsöchten Tweets wird durch die Anzahl der gesammten Benutzer in TweetsCov19 berechnet. 
df_by_username_del.describe()
```
